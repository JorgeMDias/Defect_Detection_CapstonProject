```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from pycaret.classification import *
from sklearn.model_selection import train_test_split
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Load the CSVs
paths =['/kaggle/input/binaryclassificationdatasets/binary_strat1_le_ss.csv','/kaggle/input/binaryclassificationdatasets/binary_strat1_le_ss_pca.csv','/kaggle/input/binaryclassificationdatasets/binary_strat2_le_ss.csv','/kaggle/input/binaryclassificationdatasets/binary_strat2_le_ss_pca.csv']
for path in paths:
    print("-------",path,"_________")
    df = pd.read_csv(path)
    # Split the data into training and testing datasets while stratifying on 'Defect'
    data_train_binary, data_test_binary = train_test_split(df, test_size=0.3, stratify=df['Defect'], random_state=786)
    data_unseen = data_test_binary
    exp_name = setup(data = data_train_binary,  target = 'Defect',fix_imbalance=True,fix_imbalance_method='SMOTE',data_split_stratify=True, fold_strategy='stratifiedkfold',preprocess=False,fold_shuffle=True)
  # Compare the models performance
    print(compare_models(fold = 5, cross_validation=True,sort="AUC"))
```
